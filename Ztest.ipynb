# Z test
A Z-test is a statistical hypothesis test used to determine whether there is a significant difference between a sample mean and a population mean when the population standard deviation is known or the sample size is large.

```python
import numpy as np 
from scipy.stats import norm 
sample = [172, 174, 168, 169, 171, 173, 175, 170, 169, 172]
population_mean = 170
population_std = 3
samp_mean = np.mean(sample)
n = len(sample)
```

```python
z_score = (samp_mean - population_mean) / (population_std / np.sqrt(n))
z_score
```

```python
# from statsmodels.stats.weightstats import ztest

# data = [102,104,98,101,100,
#         105,99,103,97,100]

# z_stat, p_value = ztest(sample, value=100)

# print("Z-statistic =", z_stat)
# print("P-value =", p_value)
```

```python
p_value = 2 * (1 - norm.cdf(abs(z_score)))
p_value
```

Yahan:

norm.cdf() = cumulative probability

2* isliye kyunki two-tailed test hai

```python
alpha = 0.05

if p_value < alpha:
    print("I will reject the null hypothesis")
else:
    print("i will accept the null hypothesis ")
```

# One-Sample T-Test

```python
from scipy.stats import ttest_1samp

data = [102, 104, 98, 101, 100,
        105, 99, 103, 97, 100]

t_stat, p_value = ttest_1samp(data, popmean=100)

print("T-statistic =", t_stat)
print("P-value =", p_value)
```
