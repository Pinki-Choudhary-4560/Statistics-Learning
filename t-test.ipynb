# T test
A T-test is a statistical hypothesis test used to determine whether there is a significant difference between a sample mean and a population mean (or between two sample means) when the population standard deviation is unknown.

Z-Test  |    vs  | T-Test

Feature	|   Z-Test	|    T-Test

Population Standard Deviation | 	Known| 	Unknown

Sample Size	| Large (>30)	| Small (≤30)

Distribution Used |	Normal Distribution |	t-Distribution

Test Statistic |	Z-score |	t-score

```python
import numpy as np 
from scipy import stats 
sample = [172, 174, 168, 169, 171, 173, 175, 170, 169, 172]
sample_mean = np.mean(sample)
sample_std = np.std(sample,ddof = 1)
n = len(sample)
mean_population = 170
```

```python
t_stats = (sample_mean - mean_population) / (sample_std / np.sqrt(n))
t_stats
```

```python
p_value = 2 * (1 - stats.t.cdf(abs(t_stats),df = n -1 ))
p_value
```

```python
alpha = 0.05
if p_value < alpha:
    print("reject the null hypothesis")
else:
    print("accept the null hypothesis")
```
