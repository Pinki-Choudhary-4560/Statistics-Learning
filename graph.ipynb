# Bar graph (Categorical Data Comparison)
Use: Different categories ko compare karne ke liye.

Example: Students in different courses

```python
import matplotlib.pyplot as plt

courses = ['ML', 'DL', 'AI', 'Python']
students = [50, 40, 60, 80]

plt.bar(courses, students)
plt.xlabel("Courses")
plt.ylabel("Students")
plt.title("Students by Course")
plt.show()
```

# Line Chart (Trend Over Time)
Use: Time ke saath trend dekhne ke liye.

Example: Monthly Sales

```python
import matplotlib.pyplot as plt

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
sales = [10, 15, 20, 18, 25, 30]

plt.plot(months, sales, marker='o')
plt.xlabel("Month")
plt.ylabel("Sales")
plt.title("Monthly Sales")
plt.show()
```

# Pie Chart (Part-to-Whole Relationship)
Use: Total ka kitna percentage kis category ka hai.

Example: Market Share

```python
import matplotlib.pyplot as plt

companies = ['A', 'B', 'C', 'D']
share = [40, 30, 20, 10]

plt.pie(share, labels=companies, autopct='%1.1f%%')
plt.title("Market Share")
plt.show()
```

# Ogive Curve
An Ogive is a cumulative frequency curve that shows the cumulative total of frequencies up to different class intervals.

```python
import matplotlib.pyplot as plt

months = ['Jan', 'Feb', 'Mar', 'Apr']
cum_expense = [10, 24, 42, 54]

plt.plot(months, cum_expense, marker='o')
plt.xlabel("Month")
plt.ylabel("Cumulative Expense (k)")
plt.title("Ogive (Cumulative Frequency Curve)")
plt.grid(True)

plt.show()
```

# Histogram
A Histogram is used to understand the distribution of numerical data.

```python
import matplotlib.pyplot as plt

marks = [45, 50, 55, 60, 62, 65, 70, 72, 75, 80, 85, 90]

plt.hist(marks, bins=5)

plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.title("Histogram of Marks")

plt.show()
```

# Measure of Central Tendency

A Measure of Central Tendency is a single value that represents the center or typical value of a dataset.

There are 3 main measures:

# 1. Mean (Average)

	​
Example:

Data = [10, 20, 30, 40, 50]

Mean = (10 + 20 + 30 + 40 + 50) / 5 = 30

```python
import numpy as np

data = [10, 20, 30, 40, 50]

mean = np.mean(data)
print("Mean =", mean)
```

# 2. Median (Middle Value)
Arrange data in ascending order.
Middle value = Median.

Example:

Data = [10, 20, 30, 40, 50]

Median = 30

```python
import numpy as np

data = [10, 20, 30, 40, 50]

median = np.median(data)
print("Median =", median)
```

# 3. Mode (Most Frequent Value)

Example:

Data = [10, 20, 20, 30, 40]

Mode = 20

```python
from scipy import stats

data = [10, 20, 20, 30, 40]

mode = stats.mode(data)
print("Mode =", mode.mode)
```

```python
import pandas as pd

data = pd.Series([10, 20, 20, 30, 40])

print(data.mode())
```

# Measure of Spread (Dispersion)

Measure of Spread tells us how much the data values are spread out from the center.

# 1. Range

Formula:

Range=Maximum−Minimum

```python
data = [10, 20, 30, 40, 50]

range_value = max(data) - min(data)

print("Range =", range_value)
```

# 2. Variance

Variance measures how far values are from the mean.

```python
import numpy as np

data = [10, 20, 30, 40, 50]

variance = np.var(data)

print("Variance =", variance)
```

# 3. Standard Deviation

Standard deviation is the square root of variance.

```python
import numpy as np

data = [10, 20, 30, 40, 50]

std_dev = np.std(data)

print("Standard Deviation =", std_dev)
```

# 4. Interquartile Range (IQR)

IQR measures the spread of the middle 50% of data.

Formula:

IQR=Q3-Q1​


```python
import numpy as np

data = [10, 20, 30, 40, 50]

Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)

IQR = Q3 - Q1

print("IQR =", IQR)
```
