# Two-Sample T-Test
A Two-Sample T-Test is a statistical test used to compare the means of two independent groups and determine whether the difference between them is statistically significant.

```python
import numpy as np 
from scipy import stats 
group_A = [85, 88, 90, 92, 87, 85, 89, 91, 86, 88] 
group_B = [82, 84, 80, 83, 81, 79, 78, 85, 84, 83]  
```

```python
t_stats , p_value = stats.ttest_ind(group_A, group_B , equal_var= False)
t_stats
```

```python
alpha = 0.05
if p_value < alpha:
    print("I will reject the null hypothesis ")
else:
    print("i will accept the null hypothesis" )
```

# Two-Sample Z-Test
A Two-Sample Z-Test is used to determine whether the means of two independent populations are significantly different when:

Population standard deviations are known, or
Sample sizes are large (typically n > 30).

Hypotheses:

H₀: Mean₁ = Mean₂

H₁: Mean₁ ≠ Mean₂

```python
# from statsmodels.stats.weightstats import ztest

# group1 = [85, 88, 90, 92, 87, 89, 91]
# group2 = [78, 80, 82, 79, 81, 83, 77]

# z_stat, p_value = ztest(group1, group2)

# print("Z-statistic:", z_stat)
# print("P-value:", p_value)
```

```python
import numpy as np
from scipy.stats import norm

group1 = [85, 88, 90, 92, 87, 89, 91]
group2 = [78, 80, 82, 79, 81, 83, 77]

# Means
mean1 = np.mean(group1)
mean2 = np.mean(group2)

# Sample sizes
n1 = len(group1)
n2 = len(group2)

# Standard deviations
std1 = np.std(group1, ddof=1)
std2 = np.std(group2, ddof=1)

# Z-statistic
z_stat = (mean1 - mean2) / np.sqrt((std1**2 / n1) + (std2**2 / n2))

# Two-tailed p-value
p_value = 2 * (1 - norm.cdf(abs(z_stat)))

print("Mean1 =", mean1)
print("Mean2 =", mean2)
print("Z-statistic =", z_stat)
print("P-value =", p_value)
```
